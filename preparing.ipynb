GENERAL FUNCTIONS

```python
def substring_date(a):  #for mail online to substract only date of the whole string
   return a[a.find('\n')+2:a.find(',')]
   #return a[a.find('\n\n ')+2:a.find(',')]

def month_converter(date):
       d =date.split()
    
       months = {'jan': '01','feb': '02','mar': '03','apr':'04','may':'05','jun':'06','jul':'07','aug':'08','sep':'09','oct':'10','nov':'11','dec':'12'}
       a = d[0].strip()[:3].lower()
       d[0]=months[a]
       
       if len(d[1]) <2:
            d[1]='0'+d[1]
       date_format = [d[1],d[0],d[2]]
       return "/".join(date_format)
    #    if d[0].isnumeric():
    #          a = d[1].strip()[:3].lower()
    #          if len(d[0]) <2:
    #                 d[0]='0'+d[0]
    #          d[1]=months[a]
    #          return "/".join(d)
    #    else:

        

      
def b(text):
    txt = text.split()
    
    day = txt[1]
    for ch in ['st','th','nd','rd']:
        if ch in day:
            day = day.replace(ch,"")
    txt[1]=day
    return " ".join(txt)


def remove_signs(txt):
    for ch in ['…','...','--',',','–','+','-',':',';','★','☆','/','\'@','#','%','^','&','*','(',')','=','_',']','[','“','”','‘','’','—']:
        if ch in txt:
            txt = txt.replace(ch,"")
    return txt



def add_spaces(phrase,word):  #in some articles when search keyword is changed by tags in html - need to add spaces
    #first,last,word_with_spaces="","",""
    final=[]
    as_list= phrase.split()
    for w in as_list:
        if word in w and len(w)!=len(word):
            index_first = w.find(word)
            index_last = index_first +len(word)
            first = w[:index_first]
            last = w[index_last:]
            word_with_spaces = " "+word+" "
            final.append(first+word_with_spaces+ last)
        else:
            final.append(w)
    return " ".join(final)



```
